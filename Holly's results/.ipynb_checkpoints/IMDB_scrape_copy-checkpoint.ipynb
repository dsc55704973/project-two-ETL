```python
import requests
import pandas as pd
from config import api_key
```

```python
tv_shows_path = "Resources/TV_shows.csv"

# Import the data into a Pandas DataFrame
tv_show = pd.read_csv(tv_shows_path)
tv_show
tv_show["Popularity"]=""
tv_show["ID"]=""
tv_show["TMDB Rating"]=""
df=tv_show.drop(['Unnamed: 0'], axis=1)
df
```

```python
shows=list(df["Title"])
ID=[]
```

```python
for index, row in df.iterrows():
    try:
        url = 'https://api.themoviedb.org/3/search/tv?'
        show=row['Title']      
        query_url=f'{url}api_key={api_key}&query={show}'
        response = requests.get(query_url).json()
        df.loc[index, 'ID']=response['results'][0]['id']
        df.loc[index, 'Popularity']=response['results'][0]['popularity']
        df.loc[index, 'TMDB Rating']=response['results'][0]['vote_average']

    except(KeyError, IndexError):
        print(f'No match for {show}')
df
    
```

```python
df.to_csv('TV_shows_v2.csv')
```
