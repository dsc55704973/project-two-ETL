```python
# SQLAlchemy
from sqlalchemy import create_engine
import pandas as pd
```

```python
df=pd.read_csv('..//Outputs/TV_shows_v5.csv')
df
```

```python
engine = create_engine("sqlite:///tv_data.sqlite", echo=True)
sqlite_conn = engine.connect()
```

```python
sqlite_table='TvShows'
df.to_sql(sqlite_table, sqlite_conn, if_exists='fail')
```

```python
engine.execute('SELECT * FROM TvShows LIMIT 5').fetchall()
```
